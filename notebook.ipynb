```python
#Data Cleaning
import pandas as pd
import json

df = pd.read_csv("./sentiment_analysis/Bloomberg_Transcript_Result.csv")

df["LLM Response"] = df["LLM Response"].apply(json.loads)

df
```

```python
sector_map = {
    "Information Technology": [
        "Information Technology", "information_technology", "InformationTechnology",
        "Information_Technology", "Technology", "S&P 500 Information Technology", "XLK"
    ],
    "Health Care": [
        "Health Care", "health_care", "HealthCare", "Health_Care",
        "Healthcare", "S&P 500 Health Care", "XLV"
    ],
    "Financials": [
        "Financials", "financials", "Financial",
        "S&P 500 Financials", "XLF"
    ],
    "Consumer Discretionary": [
        "Consumer Discretionary", "consumer_discretionary", "ConsumerDiscretionary",
        "Consumer_Discretionary", "Consumer Cyclical",
        "S&P 500 Consumer Discretionary", "XLY"
    ],
    "Communication Services": [
        "Communication Services", "communication_services", "CommunicationServices",
        "Communication_Services", "Communication_services",
        "S&P 500 Communication Services", "XLC"
    ],
    "Industrials": [
        "Industrials", "industrials",
        "S&P 500 Industrials", "XLI"
    ],
    "Consumer Staples": [
        "Consumer Staples", "consumer_staples", "ConsumerStaples",
        "Consumer_Staples", "Consumer Defensive",
        "S&P 500 Consumer Staples", "XLP"
    ],
    "Energy": [
        "Energy", "energy",
        "S&P 500 Energy", "XLE"
    ],
    "Utilities": [
        "Utilities", "utilities",
        "S&P 500 Utilities", "XLU"
    ],
    "Real Estate": [
        "Real Estate", "real_estate", "RealEstate",
        "Real_Estate", "Real_estate",
        "S&P 500 Real Estate", "XLRE"
    ],
    "Materials": [
        "Materials", "materials", "Basic Materials", "BasicMaterials",
        "S&P 500 Materials", "XLB"
    ]
}
```

```python
new_col = []

for data_dict in df["LLM Response"]:
    new_dict = {
        "Information Technology" : None,
        "Health Care": None,
        "Financials": None,
        "Consumer Discretionary": None,
        "Communication Services": None,
        "Industrials": None,
        "Consumer Staples" : None,
        "Energy" : None,
        "Utilities" : None,
        "Real Estate": None,
        "Materials" : None,
    }
    for key0 in new_dict:
        for key in data_dict:
            if key in sector_map[key0]:
                new_dict[key0] = data_dict[key]
                break
    new_col.append(new_dict)
            
```

```python
""" 
Some of the vals of the jsons arent floats their dictionaries, check how many of those types there
"""
correct_format_cnt = 0 #float vals
other_format_cnt = 0

incorrect_format_index = []

for index,row in enumerate(df["LLM Response"]):
    for key in row:
        if (type(row[key]) == float):
            correct_format_cnt +=1
        else:
            if (row[key] != None): 
                other_format_cnt+=1
                if index not in incorrect_format_index:
                    incorrect_format_index.append(index)

print(f"Correct Format Count: {correct_format_cnt}\nOther Format Count: {other_format_cnt}") 
```
